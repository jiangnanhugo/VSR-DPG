```python
import seaborn as sns
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import rc
# red, blue, green, yellow, purple, orange, teal blue, pink, other color
palette=['#ff3b30', '#4cd964','#ffcc00',  '#007aff','#5856d6', '#ff9500', '#5ac8fa', '#ff2d55','#969783']
markers=['o', '^', '<', 'd', '*','2']
sns.set_style("ticks")
import matplotlib as mpl
mpl.rcParams['lines.markersize'] = 6
mpl.rcParams['lines.linewidth']=1
rc("font", **{'family':'serif','serif':['Palatino'], 'size':16})
rc('text', usetex=True)
import os
```

```python
basepath="."
```

```python
def plot(total, name, title, use_logx=False, show_ylabel=True):
    GPMELD=[abs(float(line.split("\t")[0])) for line in total.split("\n")]
    PQT=[abs(float(line.split("\t")[1])) for line in total.split("\n")]
    DSR=[abs(float(line.split("\t")[2])) for line in total.split("\n")]
    VPG=[abs(float(line.split("\t")[3])) for line in total.split("\n")]
    GP=[abs(float(line.split("\t")[4])) for line in total.split("\n")]
    CVGP=[abs(float(line.split("\t")[5])) for line in total.split("\n")]
    SPL=[float(line.split("\t")[6]) for line in total.split("\n")]
    DPG=[float(line.split("\t")[7]) for line in total.split("\n")]
    
    full_data=DPG+CVGP+GP+SPL+VPG+PQT+DSR+GPMELD
    df=pd.DataFrame({
        "model": full_data,
        "type": ['VSR-DPG (ours)']*10+['VSR-GP']*10+['GP']*10+['SPL']*10+['VPG']*10+['PQT']*10+['DSR']*10+['GPMeld']*10
    })
    # df=pd.DataFrame({
    #     "model": GPMeld+PQT+DSR+VPG+GP+EGP,
    #     "type": ['GPMeld']*10+['PQT']*10+['DSR']*10+['VPG']*10+['GP']*10+['VSR-GP (ours)']*10
    # })
    

    plt.figure(figsize=(4,2.9))
    ax=sns.boxplot(data=df, x="model", y="type", 
                   showfliers=True, 
                   width=0.6,hue="type",
                    palette=palette)
    ax.set_title(title, fontsize=16)
    if use_logx==True:
        plt.xscale('log', base=10)
    ax.set(xlabel="Time Usage (Mins)", ylabel="")
    if show_ylabel==False:
        ax.set(yticklabels=[])
        
    plt.yticks(rotation=40)
    fname = os.path.join(basepath, name+"time.pdf")
    plt.savefig(fname, bbox_inches='tight', pad_inches=0)
```
